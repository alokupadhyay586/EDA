```python
state_profit=df.groupby(['State'])['Profit'].sum()
state_profit.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,8))
plt.title("State vs Profit")
state_profit.plot.bar()
```

```python
profit_sales_per_state=df.groupby('State')['Profit','Sales'].agg(['sum'])
profit_sales_per_state.plot.bar(title="Total Profit and sales per state",figsize=(15,12))
plt.xticks(rotation=90)
plt.show(block=True)
```
